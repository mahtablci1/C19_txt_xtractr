```python
#ranking algorithm
# why it keeps appending? 
rank_risk = []
import numpy as np
def ranks(list1):
    for item in  list1:
    #list =['design','paper_id','title','section','body_text']
        keywords = np.concatenate(item[0])
        rank = len(set(keywords))
        item.extend([list(set(keywords)),rank])
    return(list1)
rank_des = ranks(des_corona.copy()) 
rank_outcome = ranks(outcome_corona.copy()) 
rank_risk = ranks(risk_corona.copy()) 
```
