```python
# filter for more recent data (past 2019) that have covid in their title
import pandas as pd
import re

metadata = pd.read_csv(root+ '/output/corona_challenge/all_data_metadata.csv')
recent_data = metadata[metadata['publish_time'] >= '2019'].reset_index(drop = True)[['doc_id', 'source', 'title', 'abstract_x', 'text_body',
       'sha', 'source', 'doi', 'pmcid', 'pubmed_id', 'license', 
       'publish_time', 'authors', 'journal', 'Microsoft Academic Paper ID',
       'WHO #Covidence', 'has_full_text', 'full_text_file']]
cov_title = [title for title in recent_data['title'] if any((re.search(cov,title.lower())) for cov in cov_list)]
data = recent_data[recent_data['title'].isin(cov_title)].reset_index(drop = True)
```

```python
import re
import json
import numpy as np
risk_design_outcome =[]
smoke = []
diabete = []
pregnan = []
race = []
# smoking/diabetes / pregnancy/race
for i in range(0,len(json_filenames)):
     with open(json_filenames[i]) as file:
        data_2 = json.load(file)
        if any((re.search(cov,data_2['metadata']['title'].lower())) for cov in cov_list):
            for ii in range(0, len(data_2['body_text'])):

                risk1 = [re.findall(risk,data_2['body_text'][ii]['text']) for risk in trans_risk]
                risk1 = np.concatenate(risk1)
                if any(risk1):
                    
                    if 'smok' in risk1:
                        flag_smoke ='yes'
                        n_smoke = len([ s for s in risk1 if s == 'smok'])
                    else:
                        flag_smoke = 'no'
                        n_smoke = 0
                    title_smoke = 'no'    
                    if 'somk' in data_2['metadata']['title'].lower():
                        title_smoke = 'yes'
                        n_smoke = 10+ n_smoke
                        
                    if 'diabete'in risk1:
                        flag_diabete ='yes'
                        n_diabete = len([ s for s in risk1 if s == 'diabete'])
                    else:
                        flag_diabete = 'no'
                        n_diabete = 0
                    title_diabete = 'no'    
                    if 'diabete' in data_2['metadata']['title'].lower():
                        title_diabete = 'yes'
                        n_diabete = 10+ n_diabete
                        
                    if 'pregnan'in risk1:
                        flag_pregnan ='yes'
                        n_pregnan = len([ s for s in risk1 if s == 'pregnan'])
                    else:
                        flag_pregnan = 'no'
                        n_pregnan = 0
                    title_pregnan = 'no'    
                    if 'pregnan' in data_2['metadata']['title'].lower():
                        title_pregnan = 'yes'
                        n_pregnan = n_pregnan +10
                    
                    if ('tubercul' or 'MTB' or 'TB') in risk1:
                        flag_race ='yes'
                        n_race = len([ s for s in risk1 if s == ('tubercul' or 'MTB' or 'TB' or 'tb' or 'mtb')])
                    else:
                        flag_race = 'no'
                        n_race = 0
                    title_race = 'no'    
                    if 'tubercul' in data_2['metadata']['title'].lower():
                        title_race = 'yes'
                        n_race = n_race + 0
                        
                design = [re.findall(des,data_2['body_text'][ii]['text']) for des in trans_desi]
                design = np.concatenate(design)
                design_rank = 0
                if any(design):
                    
                    if ('meta-analysis'or 'meta analysis') in design:
                        flag_meta_analysis = 'yes'
                        design_rank = 6
                    if ('randomized controlled') in design:
                        design_rank = 5
                        flag_randomized = 'yes'
                    if ('pseudo-randomized controlled' or'pseudo randomized controlled') in design:
                        design_rank = 4
                        flag_pseudo_randomized = 'yes'
                    design_rank = design_rank + len(design)  
                    
                outcome_rank = 0
                outcome = [re.findall(des,data_2['body_text'][ii]['text']) for des in trans_outcom]
                outcome = np.concatenate(outcome)
                if any(outcome):
                    outcome_rank = len(outcome)
                
                if flag_smoke == 'yes':
                    smoke.append([data_2['paper_id'],data_2['metadata']['title']
                                       ,data_2['body_text'][ii]['section'], data_2['body_text'][ii]['text']
                                       ,risk1, title_smoke, n_smoke
                                       , design, outcome, outcome_rank,design_rank,data_2['body_text']])
            
                if flag_diabete == 'yes':
                    diabete.append([data_2['paper_id'],data_2['metadata']['title']
                                       ,data_2['body_text'][ii]['section'], data_2['body_text'][ii]['text']
                                       ,risk1, title_diabete, n_diabete
                                      , design, outcome, outcome_rank,design_rank,data_2['body_text']])
                if flag_pregnan == 'yes':
                    pregnan.append([data_2['paper_id'],data_2['metadata']['title']
                                       ,data_2['body_text'][ii]['section'], data_2['body_text'][ii]['text']
                                       ,risk1, title_pregnan, n_pregnan
                                      , design, outcome, outcome_rank,design_rank, data_2['body_text']]) 
                if flag_race == 'yes':
                    race.append([data_2['paper_id'],data_2['metadata']['title']
                                       ,data_2['body_text'][ii]['section'], data_2['body_text'][ii]['text']
                                       ,risk1, title_race, n_race
                                      , design, outcome, outcome_rank,design_rank]) 
                            
                        
                risk_design_outcome.append([data_2['paper_id'],data_2['metadata']['title']
                                       ,data_2['body_text'][ii]['section'], data_2['body_text'][ii]['text']
                                       ,risk1, flag_smoke, title_smoke, n_smoke, flag_diabete, title_diabete, n_diabete
                                       , flag_pregnan, title_pregnan, n_pregnan, flag_race, title_race, n_race
                                      , design, outcome, outcome_rank,design_rank] ) 
                
            
```

```python
# data for smoke:

smoke_pd = pd.DataFrame(smoke, columns = ['paper_id','title','section','text','risk1'
                               , 'title_smoke', 'n_smoke', 'design', 'outcome', 'outcome_rank','design_rank'])
smoke_pd['rank'] = smoke_pd['n_smoke'] + smoke_pd['outcome_rank']/10 + smoke_pd['design_rank']
selected_smoke = smoke_pd.sort_values('rank', ascending = False).reset_index(drop = True)[0:10]
```
